```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("heart.csv")
```

```python
# age: Age of the patient.
# sex: Gender (1 = male, 0 = female).
# cp: Chest pain type (various types encoded as integers).
# trestbps: Resting blood pressure.
# chol: Serum cholesterol level.
# fbs: Fasting blood sugar (1 = True, 0 = False).
# restecg: Resting electrocardiographic results.
# thalach: Maximum heart rate achieved.
# exang: Exercise-induced angina (1 = yes, 0 = no).
# oldpeak: ST depression induced by exercise relative to rest.
# slope: The slope of the peak exercise ST segment.
# ca: Number of major vessels (0-3) colored by fluoroscopy.
# thal: Thalassemia (a blood disorder).
# target: The target variable (1 = presence of heart disease, 0 = absence).
```

```python
#1. What is the average age of patients with heart disease?
heart_disease_patients = df[df['target'] == 1]
avg_age = heart_disease_patients['age'].mean()
# Create a pie chart showing the average age
plt.figure(figsize=(6, 6))
plt.pie([avg_age, 100 - avg_age], labels=['Average Age', 'Rest'], autopct='%1.1f%%', colors=['coral', 'lightgray'])
plt.title('Average Age of Patients with Heart Disease')
plt.show()
```

```python
#2. Is there a relationship between cholesterol levels and heart disease?
avg_chol_heart_disease = df[df['target'] == 1]['chol'].mean()
avg_chol_no_heart_disease = df[df['target'] == 0]['chol'].mean()
#Violin plot to show density of data
plt.figure(figsize=(10, 6))
sns.violinplot(x='target', y='chol', data=df, palette={0: 'skyblue', 1: 'coral'})
plt.xlabel('Heart Disease Status')
plt.ylabel('Cholesterol Level')
plt.title('Cholesterol Levels Distribution by Heart Disease Status')
plt.xticks([0, 1], ['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
#3. Which chest pain type (cp) is more common among heart disease patients?
chest_pain_counts = heart_disease_patients['cp'].value_counts()
#Horizontal bar plot
plt.figure(figsize=(8, 6))
chest_pain_counts.plot(kind='barh', color='coral')
plt.xlabel('Count')
plt.ylabel('Chest Pain Type')
plt.title('Distribution of Chest Pain Types Among Heart Disease Patients')
plt.show()
```

```python
#4. Does gender affect the likelihood of heart disease?
heart_disease_by_gender = heart_disease_patients['sex'].value_counts(normalize=True)
#Pie chart 
plt.figure(figsize=(6, 6))
plt.pie(heart_disease_by_gender, autopct='%1.1f%%', labels = ['Female' if x == 0 else 'Male' for x in heart_disease_by_gender.index], 
        colors=plt.cm.Paired(range(len(heart_disease_by_gender))))
plt.title('Proportion of Heart Disease Patients by Gender')
plt.show()
```

```python
#5. What is the relationship between age and maximum heart rate achieved (thalach)?
fg=df.plot.scatter(x='age', y='thalach')
plt.show()
```

```python
#6. Does high blood pressure (trestbps) correlate with the presence of heart disease?
avg_bp_heart_disease = df[df['target'] == 1]['trestbps'].mean()
avg_bp_no_heart_disease = df[df['target'] == 0]['trestbps'].mean()
#Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='target', y='trestbps', data=df, palette='coolwarm')
plt.xlabel('Heart Disease')
plt.ylabel('Blood Pressure')
plt.title('Distribution of Blood Pressure by Heart Disease')
plt.xticks(ticks=[0, 1], labels=['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
#7. How does exercise-induced angina (exang) relate to heart disease?
angina_heart_disease = heart_disease_patients['exang'].value_counts(normalize=True)
#Pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    angina_heart_disease,
    labels=['No Angina', 'Angina'],
    autopct='%1.1f%%',
    colors=plt.cm.Paired(range(len(angina_heart_disease))),
    startangle=140
)
plt.title('Proportion of Exercise-Induced Angina Among Heart Disease Patients')
plt.show()
```
